```python
df['Gender'] = df['Gender'].astype("category")
```

```python
df['Gender'].unique()
```

```python
sns.pairplot(data = df[["Age","Annual Income (k$)","Spending Score (1-100)"]])
```

```python
sns.heatmap(data=df[["Age","Annual Income (k$)","Spending Score (1-100)"]].corr(), annot= True, cmap='coolwarm', linewidths=5, linecolor='black', vmin=0.5,vmax=1,cbar=False)
```

```python
sns.relplot(data = df, x='Annual Income (k$)', y= 'Spending Score (1-100)', hue='Gender', kind="line")
```

```python
sns.relplot(data = df, x='Age', y= 'Spending Score (1-100)', hue='Gender', kind="line")
```

```python
sns.relplot(data = df, x='Age', y= 'Annual Income (k$)', hue='Gender', kind="line")
```

```python
sns.boxplot(data=df, x ='Spending Score (1-100)')
```

```python
sns.distplot(df['Spending Score (1-100)'], hist=True, kde=True, color = 'orange')
```

```python
sns.boxplot(data=df, x ='Annual Income (k$)')
```

```python
sns.distplot(df['Annual Income (k$)'], hist=True, kde=True, color = 'orange')
```
